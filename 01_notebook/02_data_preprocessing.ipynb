```python
# =================================================================
# Hücre 1: Kurulum ve Veri Yükleme
# =================================================================
import pandas as pd
import numpy as np

# Yolunu senin projene göre ayarla
DATA_PATH = "../data/raw/gps_locations.csv"

df = pd.read_csv(DATA_PATH)

print("[OK] Veri yüklendi:", DATA_PATH)
print("Satır:", len(df), "| Sütun:", df.shape[1])
```

```python
# Hücre 2: Veri Ön İncelemesi

print("\n--- İlk 5 Satır ---")
display(df.head())

print("\n--- Veri Bilgisi ---")
print(df.info())

print("\n--- Eksik Değerler ---")
print(df.isna().sum())

print("\n--- Özet İstatistikler ---")
display(df.describe())
```

### Sonuç & Öneri
- Veri seti başarıyla yüklendi (120.000 satır, 26 sütun).  
- Eksik değer bulunmamaktadır.  
- Veri tipleri büyük ölçüde uygun görünüyor; yalnızca `timestamp` sütunu `object` tipinde ve dönüştürülmesi gerekiyor.  
- Aykırı değer olarak yüksek hız (365 km/h) gözlemlendi, ileride işaretlenmesi yararlı olacaktır.  

➡️ Sonraki adım: `timestamp` sütununun `datetime` tipine dönüştürülmesi ve zaman bazlı özniteliklerin türetilmesi (gün, saat, vb.).  


```python
# =========================================================
# H3: Zaman Damgası Dönüşümü + Bayraklar (tek sabit ile)
# Not: Temizlik YAPMAZ; sadece işaretler (flag).
# =========================================================

# -- Tek yerden yönetilen hız eşiği (km/s)
if "SPEED_CEILING_KMH" not in globals():
    SPEED_CEILING_KMH = 200  # ihtiyaca göre 180/220 olarak güncellenebilir

# 1) timestamp -> datetime
df["timestamp"] = pd.to_datetime(df["timestamp"], errors="coerce")

# 2) zaman türevleri
df["date"] = df["timestamp"].dt.date
df["hour"] = df["timestamp"].dt.hour.astype("Int64")  # geçici Int64, H4'te istersek int32'ye çeviririz
df["day_of_week"] = df["timestamp"].dt.day_name()

# 3) kontrol bayrakları
df["is_ts_missing"]   = df["timestamp"].isna().astype(int)
df["is_gps_valid"]    = (
    df["latitude"].between(-90, 90, inclusive="both")
    & df["longitude"].between(-180, 180, inclusive="both")
).astype(int)
df["is_speed_suspect"] = (df["speed"] >= SPEED_CEILING_KMH).astype(int)

# 4) rapor
n = len(df)
bad_ts = int(df["is_ts_missing"].sum())
bad_gps = int((df["is_gps_valid"] == 0).sum())
sus_spd = int(df["is_speed_suspect"].sum())

print("=== Dönüşüm & Bayrak Özeti (H3) ===")
print(f"Toplam satır: {n:,}")
print(f"Geçersiz timestamp (NaT): {bad_ts}")
print(f"Geçersiz GPS (aralık dışı): {bad_gps}")
print(f"Şüpheli hız (≥ {SPEED_CEILING_KMH} km/h): {sus_spd}")

print("\n[Zaman aralığı]")
print("Başlangıç:", df["timestamp"].min())
print("Bitiş   :", df["timestamp"].max())
```

**H3 – Kısa, net özet**

* **Veri Durumu:** Veri, zaman ve konum açısından sağlam. Eksik veya hatalı zaman damgası (NaT) ve aralık dışı GPS verisi bulunmuyor.
* **Anormallik:** Yalnızca hız tarafında 183 aşım var (≥ 200 km/h, ~%0.15). Bu anormallikler, bir sonraki adımda temizlenecek şekilde bayraklandı.
* **Aksiyon:** Veri, bir sonraki aşama olan yıkıcı temizlik (satır silme) için uygun zeminde. Bu aşamada sadece bayraklama yapıldığı için, ham verinin bir kopyası korunmaktadır.

```python
# =========================================================
# H4: Temizleme Kuralları (bayraklara göre drop + rapor)
# - H3'te üretilen bayrakları kullanır
# - Hız eşiği H3 ile AYNI sabitten alınır: SPEED_CEILING_KMH
# - Drop öncesi en hızlı 10 kaydı loglar (denetim izi)
# =========================================================

import pandas as pd

# -- Temizlik kuralları
ENFORCE_TS_VALID   = True   # NaT olan timestamp satırlarını at
ENFORCE_GPS_RANGE  = True   # -90/90, -180/180 dışını at
ENFORCE_SPEED_CEIL = True   # is_speed_suspect==1 olanı at

# -- Denetim izi: drop ÖNCESİ en hızlı 10 kayıt (gözlem için)
print("\n[Ön Kontrol] En hızlı 10 kayıt (drop öncesi):")
cols_preview = ["timestamp", "vehicle_id", "driver_id", "speed", "latitude", "longitude"]
print(df.sort_values("speed", ascending=False)[cols_preview].head(10))

# -- Silme maskesi
to_drop_mask = pd.Series(False, index=df.index)
if ENFORCE_TS_VALID:
    to_drop_mask |= (df["is_ts_missing"] == 1)
if ENFORCE_GPS_RANGE:
    to_drop_mask |= (df["is_gps_valid"] == 0)
if ENFORCE_SPEED_CEIL:
    to_drop_mask |= (df["is_speed_suspect"] == 1)

# -- Neden bazında sayımlar (rapor)
reason_counts = {
    "invalid_timestamp": int((df["is_ts_missing"] == 1).sum()),
    "invalid_gps":       int((df["is_gps_valid"] == 0).sum()),
    f"speed_≥{SPEED_CEILING_KMH}": int((df["is_speed_suspect"] == 1).sum()),
}

n_before = len(df)
n_drop   = int(to_drop_mask.sum())
n_after  = n_before - n_drop

print("\n=== Temizleme Özeti (H4) ===")
print(f"Önceki satır sayısı : {n_before:,}")
print(pd.DataFrame.from_dict(reason_counts, orient="index", columns=["silinecek_satır"]))
print(f"Toplam silinecek    : {n_drop:,}")

# -- Gerçek drop
if n_drop > 0:
    df = df.loc[~to_drop_mask].copy()
    df.reset_index(drop=True, inplace=True)

print(f"Sonraki satır sayısı : {len(df):,}")

# -- Tür optimizasyonu: kategorikler
print("\n--- Kategorik Değişkenler Dönüştürülüyor ---")
categorical_features = [
    "driver_id", "vehicle_id",
    "weather_conditions", "road_type", "traffic_condition",
    "day_of_week"
]
for col in categorical_features:
    if col in df.columns:
        df[col] = df[col].astype("category")

# hour'ı kompakt tiplendir (opsiyonel)
if "hour" in df.columns:
    df["hour"] = df["hour"].astype("int32")

# -- Hızlı kontrol
print("\n[Kontrol] Kısaltılmış info:")
print(df[["timestamp","latitude","longitude","speed"]].info())
```

**H4 – Kısa, net özet**

* **Temizleme:** Hız eşiğini aşan 183 satır (`is_speed_suspect == 1`) temizlendi. Zaman ve GPS kaynaklı kayıp yaşanmadı.
* **Veri Boyutu:** Toplam satır sayısı 120,000'den 119,817'ye düştü.
* **Optimizasyon:** `driver_id`, `vehicle_id`, `weather_conditions`, `road_type`, `traffic_condition` gibi kimlik ve bağlamsal sütunlar, bellek kullanımını azaltmak ve grup işlemlerini hızlandırmak amacıyla `category` tipine dönüştürüldü.
* **Durum:** Veri, sonraki adım olan **Yolculuk Segmentasyonu** ve **Özellik Mühendisliği** için temiz ve optimize edilmiş durumda.

```python
# =========================================================
# Hücre 5: trip_id Tekillik Kontrolü (EDA → Preprocessing köprüsü)
# Veri Seti Kontrolü: trip_id Tekil Mi?
# =========================================================

import pandas as pd

# df'in güncel durumunu kontrol et
total_rows = len(df)
unique_trip_ids = df['trip_id'].nunique()

print(f"Toplam satır sayısı: {total_rows:,}")
print(f"Tekil trip_id sayısı: {unique_trip_ids:,}")

if total_rows == unique_trip_ids:
    print("\n[Uyarı] Her bir GPS kaydı, ayrı bir yolculuk olarak tanımlanmış.")
    print("Yolculuk segmentasyonunu yeniden yapmamız gerekiyor.")
else:
    print("\n[Bilgi] Yolculuklar doğru gruplandırılmış.")
```

**Değerlendirme – Trip ID Kontrolü**  
- Toplam satır: 119,817 | Tekil trip_id: 119,817  
- Sonuç: Her satır için farklı `trip_id` atanmış. Yani veri seti, her GPS noktasını ayrı yolculuk gibi tanımlıyor.  
- Etki: Mevcut `trip_id` sütunu yolculuk segmentasyonu için kullanılamaz. Bu nedenle kendi segmentasyonumuzu zaman farkı kuralı ile yeniden tanımlamamız gerekiyor.  

```python
# =========================================================
# H6: Yolculuk Segmentasyonu + Temiz Veriyi Kaydetme
# Amacı: Yolculuk segmentlerini oluşturmak ve temiz df'yi kaydetmek
# =========================================================

import os
import datetime as dt
import pandas as pd

# Güvenlik: df var mı?
if "df" not in globals():
    raise RuntimeError("df bulunamadı. Önceki hücreleri çalıştırdığınızdan emin olun.")

print("=== Yolculuk Segmentasyonu (H6) – araç bazında ===")

# --- Parametreler
# 5 dakika (300 saniye) ve 10 dakika (600 saniye) yaygın olarak kullanılır.
# 5 dakika, dur-kalk veya kısa molalar için daha uygun olabilir.
GAP_SEC = 300       # 5 dakika
MIN_ROWS_PER_SEG = 2 # en az 2 satır şartı (tek nokta segmentleri eleyelim)

# 1) Güvenli sıralama: önce vehicle_id, sonra timestamp
df.sort_values(by=['vehicle_id', 'timestamp'], inplace=True)

# 2) Araç bazında ardışık zaman farkı (saniye)
df['time_diff_sec'] = df.groupby('vehicle_id')['timestamp'].diff().dt.total_seconds().fillna(0)

# 3) Yeni segment bayrağı: araç değişimi veya zaman boşluğu > GAP_SEC
df['new_seg'] = ((df['vehicle_id'] != df['vehicle_id'].shift()) | (df['time_diff_sec'] > GAP_SEC)).astype(int)

# 4) Araç içi kümülatif toplama ile yerel segment id'si
df['trip_seg_local'] = df.groupby('vehicle_id')['new_seg'].cumsum()

# 5) Global benzersiz segment id
df["trip_seg_id"] = pd.factorize(df['vehicle_id'].astype(str) + '_' + df['trip_seg_local'].astype(str))[0] + 1

# Segment boyutlarına göre filtreleme
seg_counts = df.groupby('trip_seg_id').size()
valid_segments = seg_counts[seg_counts >= MIN_ROWS_PER_SEG].index
df = df[df['trip_seg_id'].isin(valid_segments)].copy()

print(f"Toplam segment sayısı (min {MIN_ROWS_PER_SEG} satır): {df['trip_seg_id'].nunique():,}")
print(f"Segmentasyondan sonraki satır sayısı: {len(df):,}")

print("\n--- Temiz Seti Kaydetme ---")
# --- Kayıt yolları
CLEAN_DIR = "../data/clean"
BASE_NAME = "gps_locations_clean.csv"
TS = dt.datetime.now().strftime("%Y%m%d_%H%M%S")
VERSIONED_NAME = f"gps_locations_clean_{TS}.csv"
os.makedirs(CLEAN_DIR, exist_ok=True)
base_path = os.path.join(CLEAN_DIR, BASE_NAME)
ver_path  = os.path.join(CLEAN_DIR, VERSIONED_NAME)

# --- Kaydet
df.to_csv(base_path, index=False)
df.to_csv(ver_path,  index=False)

print(f"[OK] Ana dosya  : {base_path}")
print(f"[OK] Sürüm dosya: {ver_path}")
print(f"Satır/Sütun     : {df.shape[0]:,} x {df.shape[1]}")
```
